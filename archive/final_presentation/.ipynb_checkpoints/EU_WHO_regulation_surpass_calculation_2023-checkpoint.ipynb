```python
import pandas as pd

# Import the CSV and automatically parse the datetime column
df = pd.read_csv('../data/fine_dust_complete.csv', parse_dates=['datetime'])

# Step 1: Filter data for the year 2023
df_2023 = df[df['datetime'].dt.year == 2023].copy()

# Step 2: Calculate the yearly average for 2023 (WHO Standard)
yearly_average_2023 = df_2023['value'].mean()

# Step 3: Group by day to calculate daily averages
df_2023.loc[:, 'date'] = df_2023['datetime'].dt.date  # Extract the date part
df_2023.loc[:, 'weekday'] = df_2023['datetime'].dt.day_name()  # Extract the weekday name

# Group by date and weekday to calculate daily averages
daily_averages_2023 = df_2023.groupby(['date', 'weekday'])['value'].mean().reset_index()

# Step 4: Filter the days where the daily average exceeded 35 µg/m³ (EU Standard)
exceedances_2023_df = daily_averages_2023[daily_averages_2023['value'] > 35]

# Ensure the resulting dataframe is ordered by datetime
exceedances_2023_df = exceedances_2023_df.sort_values(by='date').reset_index(drop=True)

# Display the yearly average and the dataframe of exceedances
exceedances_2023_df.head(60)
```

```python
# Step 2: Filter data for March 1st, 2023
march_first_df = df_2023[(df_2023['datetime'].dt.month == 3) & (df_2023['datetime'].dt.day == 1)].copy()

# Extract the hour from datetime for plotting
march_first_df['hour'] = march_first_df['datetime'].dt.hour

# Plotting the PM10 values for March 1st, 2023
plt.figure(figsize=(12, 6))
sns.lineplot(data=march_first_df, x='hour', y='value', errorbar=None)
plt.title('PM10 Values on March 1st, 2023')
plt.xlabel('Hour')
plt.ylabel('PM10 Value')
plt.xticks(range(0, 24, 1))  # Set x-ticks for every hour
plt.tight_layout()

# Show the plot
plt.show()
```
