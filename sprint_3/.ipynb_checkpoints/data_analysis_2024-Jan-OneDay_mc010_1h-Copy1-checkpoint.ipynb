```python
start_time = pd.to_datetime("08:00:00").time()  # Set start time to 8 am
end_time = pd.to_datetime("18:00:00").time()    # Set end time to 6 pm

# Extract the time component from the datetime
df["time"] = df["datetime"].dt.time

# Filter rows where "time" is between start_time and end_time
df_filtered_working_hours = df[(df["time"] >= start_time) & (df["time"] <= end_time) & (df["core"] == "pm10") ]

df_filtered_working_hours.count()
```

```python
target_day = '2024-01-13'
df_filtered_day = df_pm10[df_pm10['datetime'].dt.date == pd.to_datetime(target_day).date()]
```

```python
#create new columns hour with map
def cutDatetimeToHour(datetime):
    return datetime.hour

df_filtered_day["hour"] = df_filtered_day["datetime"].map(cutDatetimeToHour)
df_filtered_day.head(24)
```

```python
plt.plot(df_filtered_day['hour'], df_filtered_day['value'], color="blue", linewidth=1)
plt.xlabel('Hour of Day')
plt.ylabel('pm10 value')
plt.title(f'Plot of Value on {target_day}')
plt.xticks(df_filtered_day['hour'])
plt.show
```

```python
target_day = '2024-01-14'
df_filtered_day2 = df_pm10[df_pm10['datetime'].dt.date == pd.to_datetime(target_day).date()]
#create new columns hour with map
def cutDatetimeToHour(datetime):
    return datetime.hour

df_filtered_day2["hour"] = df_filtered_day2["datetime"].map(cutDatetimeToHour)
#df_filtered_day2.head(24)
plt.plot(df_filtered_day2['hour'], df_filtered_day2['value'], color="blue", linewidth=1)
plt.xlabel('Hour of Day')
plt.ylabel('pm10 value')
plt.title(f'Plot of Value on {target_day}')
plt.xticks(df_filtered_day2['hour'])
plt.show
```
